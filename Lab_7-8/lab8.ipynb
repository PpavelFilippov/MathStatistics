```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import os
os.chdir("/Users/pavelfilippov/Downloads")
```

```python
def read(filename):
    col1 = []
    col2 = []
    col3 = []

    with open(filename, 'r') as file:
        for line in file:
            elements = line.split()
            if len(elements) >= 4 :
                col1.append(float(elements[1]))
                col2.append(float(elements[2]))
                col3.append(float(elements[3]))
    return np.array(col1), np.array(col2), np.array(col3)                
```

```python
filename = 'data.dat'
sample1, sample2, sample3 = read(filename)
print(sample1)

```

```python
def fisher_test(x, y, alpha=0.05):
  var_x = np.var(x, ddof=1)
  var_y = np.var(y, ddof=1)

  F = var_x / var_y

  k_x = len(x) - 1
  k_y = len(y) - 1

  F_critical = scipy.stats.f.ppf(1 - alpha / 2, k_x, k_y)

  return var_x, var_y, F, k_x, k_y, F_critical, F < F_critical
```

```python
fisher_test(sample1, sample2)
```

```python
fisher_test(sample1, sample1[205:410])
```

```python
fisher_test(sample1[0:205], sample1[205:615])
```
